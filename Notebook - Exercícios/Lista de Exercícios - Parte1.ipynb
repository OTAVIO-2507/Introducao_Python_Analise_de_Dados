13. Crie um código que calcule e imprima a média ponderada dos números 5, 12, 20 e 15 com pesos respectivamente iguais a 1, 2, 3 e 4.

```python
n1 = 5
p1 = 1

n2 = 12
p2 = 2

n3 = 20
p3 = 3

n4 = 15
p4 = 4

media_ponderada = ((n1*p1) + (n2*p2) + (n3*p3) + (n4*p4)/(p1+p2+p3+p4))

print(f'O valor da média ponderada é igual a {media_ponderada}')
```
